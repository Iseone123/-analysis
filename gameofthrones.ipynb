```python
cat_columns = ['title', 'culture', 'mother', 'father', 'heir', 'house', 'spouse']
df[cat_columns] = df[cat_columns].fillna('Unknown')
```

```python
df['dateOfBirth'] = df['dateOfBirth'].fillna(df['dateOfBirth'].median())
df['age'] = df['age'].fillna(df['age'].median())
```

```python
alive_cols = ['isAliveMother', 'isAliveFather', 'isAliveHeir', 'isAliveSpouse']
df[alive_cols] = df[alive_cols].fillna(0.0)
```

```python
df
```

```python
def popularity_s(popularity):
    if popularity<0.5:
        return 0
    else:
        return 1
```

```python
df["isPopular"]=df["popularity"].apply( popularity_s)
```

```python
df
```

```python
df['boolDeadRelations']=df['numDeadRelations'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df
```

```python
df['culture'] = df['culture'].str.lower()
```

```python
df['culture']
```

```python
cult = {
    'Summer Islands': ['summer islands', 'summer islander', 'summer isles'],
    'Ghiscari': ['ghiscari', 'ghiscaricari',  'ghis'],
    'Asshai': ["asshai'i", 'asshai'],
    'Lysene': ['lysene', 'lyseni'],
    'Andal': ['andal', 'andals'],
    'Braavosi': ['braavosi', 'braavos'],
    'Dornish': ['dornishmen', 'dorne', 'dornish'],
    'Myrish': ['myr', 'myrish', 'myrmen'],
    'Westermen': ['westermen', 'westerman', 'westerlands'],
    'Westerosi': ['westeros', 'westerosi'],
    'Stormlander': ['stormlands', 'stormlander'],
    'Norvoshi': ['norvos', 'norvoshi'],
    'Northmen': ['the north', 'northmen'],
    'Free Folk': ['wildling', 'first men', 'free folk'],
    'Qartheen': ['qartheen', 'qarth'],
    'Reach': ['the reach', 'reach', 'reachmen'],
}


```

```python
culture_map = {}
for group, aliases in cult.items():
    for alias in aliases:
        culture_map[alias] = group
df['culture'] = df['culture'].apply(lambda x: culture_map.get(x, x) if pd.notna(x) else x)
```
