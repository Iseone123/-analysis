```python
df=df.drop(columns=('Flavanoids'))
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df.columns, index=df.index)
```

```python
scaled_df
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.cluster import KMeans

sse = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(scaled_df)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, sse, 'bo-')
plt.xlabel('Количество кластеров k')
plt.ylabel('SSE (Inertia)')
plt.title('Метод локтя')
plt.show()
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# 3. Кластеризация KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(pca_data)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1],
            c=kmeans.labels_, cmap='plasma', s=40)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Центры кластеров')
plt.title("K-Means с PCA (2 компоненты)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.colorbar(label='Кластер')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Без PCA
kmeans_full = KMeans(n_clusters=3, random_state=42).fit(scaled_data)
score_full = silhouette_score(scaled_data, kmeans_full.labels_)

# С PCA
kmeans_pca = KMeans(n_clusters=3, random_state=42).fit(pca_data)
score_pca = silhouette_score(pca_data, kmeans_pca.labels_)

print("Silhouette без PCA:", score_full)
print("Silhouette с PCA:", score_pca)
```
