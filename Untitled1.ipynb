```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("superhero_movies_dataset.csv")
```

```python
df.fillna(0)
```

```python
df[['Release_Type', 'Worldwide_Gross']] = df['Worldwide_Gross'].str.extract(r'([^\(]+)\s*\(\$(.*?)\)')
```

```python
df[df['Worldwide_Gross'].str.count('\.') > 1]
```

```python
df = df.drop(185)
```

```python
df[df['Worldwide_Gross'].str.count('\.') > 1]
```

```python
df['Worldwide_Gross'] = df['Worldwide_Gross'].str.replace(',', '').astype(float)
```

```python
df=df.dropna()
```

```python
df.groupby('Distributor',as_index=False).agg({'Worldwide_Gross':'sum'}).sort_values('Worldwide_Gross',ascending=False)
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
```

```python
df.groupby('Year')['Worldwide_Gross'].sum().plot()
```

```python
df.groupby("Publisher").agg({"Film":"count"}).sort_values("Film",ascending=False)
```

```python
df.sort_values('Worldwide_Gross', ascending=False).head(10)
```

```python
df['Worldwide_Gross']=df['Worldwide_Gross'].apply(lambda x: np.sqrt(x))
```

```python
df['Worldwide_Gross'].hist(bins=50)
```

```python
df.groupby('Year')['Worldwide_Gross'].sum().sort_index().plot(kind='line', figsize=(10,5), title="Суммарные сборы по годам")
plt.xlabel("Год")
plt.ylabel("Сборы ($)")
plt.grid(True)
plt.show()
```

```python
top_distributors = df.groupby('Distributor')['Worldwide_Gross'].sum().sort_values(ascending=False).head(10)
top_distributors.plot(kind='bar', figsize=(10,5), title="Топ-10 дистрибьюторов по сборам")
plt.ylabel("Суммарные сборы ($)")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Worldwide_Gross'].hist(bins=50, figsize=(10,5))
plt.title("Распределение сборов")
plt.xlabel("Сборы ($)")
plt.ylabel("Количество фильмов")
plt.show()
```

Наибольшие сборы показывают такие компании, как Paramount Pictures, Universal Pictures и Sony Pictures.
